```python
import pandas as pd
diabetes_data = pd.read_csv('diabetes_prediction_dataset.csv')
diabetes_data
```

```python
#to check the missing values
if diabetes_data.isnull().values.any():
    print("There are missing values in the DataFrame.")
else:
    print("There are no missing values in the DataFrame.")
```

```python
#to remove the duplicate values 
diabetes_data.drop_duplicates(inplace=True)
diabetes_data
```

```python
diabetes_data['gender'] = diabetes_data['gender'].replace({'Male': 1, 'Female': 0})
print(diabetes_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Create a pairplot for the entire dataset
sns.pairplot(diabetes_data)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = diabetes_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
diabetes_data.hist(bins=20, figsize=(12, 10))
plt.suptitle("Histograms of Numerical Columns")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
diabetes_data['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Gender')
plt.show()
```

```python
import seaborn as sns 
plt.figure(figsize=(12, 6))
sns.boxplot(data=diabetes_data.drop(columns=['gender']))
plt.title('Box Plot of Numerical Columns')
plt.show()
```
